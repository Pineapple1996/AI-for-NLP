```python
def search(start,end,city_dict):
    pathes = [[start]]
    visited = set()
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]
        if frontier in visited:continue
        con = city_dict[frontier]
        for sta in con:
            if sta in path : continue
            new = path +[sta]
            pathes.append(new)
            visited.add(sta)
            if sta== end:return new
            print(new)
```

```python
search('北京','上海',city_dict)
```

$$ loss = \frac {1}{n} \sum(yi-\hat{yi})$$

```python
def loss(y,y_h):
    return  sum((y_i - y_h_i)**2 for y_i, y_h_i in zip(list(y), list(y_h)))/len(list(y))
```
