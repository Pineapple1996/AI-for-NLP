```python
import re
source_list = re.findall('\{name:\'(.*)\', geoCoord:\[(.*), (.*)\]\}',coordination_source)
source_list
```

```python
city_dict = {}
for i in source_list:
    city_dict[i[0]]=(float(i[1]),float(i[2]))
city_dict
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
city_graph = nx.Graph()
print(list(city_dict.keys()))
city_graph.add_nodes_from(list(city_dict.keys()))
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] 
matplotlib.rcParams['font.family']='sans-serif'
nx.draw(city_graph, city_dict, with_labels=True, node_size=30)
```

```python
def search(start,end,city_dict):
    pathes = [[start]]
    visited = set()
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]
        if frontier in visited:continue
        con = city_dict[frontier]
        for sta in con:
            if sta in path : continue
            new = path +[sta]
            pathes.append(new)
            visited.add(sta)
            if sta== end:return new
            print(new)
```

```python
search('北京','上海',city_dict)
```
